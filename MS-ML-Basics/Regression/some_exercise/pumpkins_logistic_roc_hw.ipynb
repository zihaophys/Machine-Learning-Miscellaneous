```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
```

```python
pumpkins
```

```python
from sklearn.pipeline import Pipeline
from sklearn.pipeline import make_pipeline
```

```python
cat_features = [0, 1, 2, 5]
num_features = [3, 4]
```

```python
numeric_transformer = make_pipeline(StandardScaler())
categorical_transformer = make_pipeline(OneHotEncoder())
```

```python
from sklearn.compose import ColumnTransformer
```

```python
preprocessor = ColumnTransformer(transformers=[('num', numeric_transformer, num_features), ('cat', categorical_transformer, cat_features)])
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
pipline = make_pipeline(preprocessor, LogisticRegression(C=100, solver='liblinear'))
```

```python
from sklearn.preprocessing import LabelEncoder
pumpkins.Color = pumpkins[['Color']].apply(LabelEncoder().fit_transform)
```

```python
pumpkins['Color']
```

```python
from sklearn.model_selection import train_test_split
X, y = pumpkins[pumpkins.columns[:-1]].values, pumpkins.Color.values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
pipline.fit(X_train, y_train)
```

```python
y_pred = pipline.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
```

```python
accuracy_score(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

```python
print(classification_report(y_train, pipline.predict(X_train)))
```

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import confusion_matrix

y_scores = pipline.predict_proba(X_test)


fpr, tpr, thresholds = roc_curve(y_test, y_scores[:,1])


fig = plt.figure(figsize=(6, 6))

plt.plot([0, 1], [0, 1], 'k--')

plt.plot(fpr, tpr)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.show()
```
