```python
import pandas as pd
```

```python
pumpkins = pd.read_csv('US-pumpkins.csv')
pumpkins.head()
```

```python
pumpkins.tail()
```

```python
pumpkins['Month'] = pd.DatetimeIndex(pumpkins['Date']).month
```

```python
pumpkins.head()
```

```python
new_columns = ['Package', 'Month', 'Low Price', 'High Price', 'Date']
pumpkins = pumpkins[new_columns]
```

```python
list(pumpkins.columns)
```

```python
pumpkins.isnull().sum()
```

```python
price = (pumpkins['Low Price'] + pumpkins['High Price'])/2
month = pumpkins['Month']
```

```python
pumpkins['Price'] = price
```

```python
pumpkins.head()
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(9,6))
ax = fig.gca()
pumpkins.boxplot('Price', by='Month', ax=ax)
plt.show()
```

```python
new_pumpkins = pd.DataFrame({'Month': month, 'Package': pumpkins['Package'], 'Low Price': pumpkins['Low Price'], 'High Price': pumpkins['High Price'], 'Price': price})
```

```python
new_pumpkins.head()
```

```python
new_pumpkins
```

```python
new_pumpkins = new_pumpkins[new_pumpkins['Package'].str.contains('bushel')]
```

```python
new_pumpkins
```

```python
new_pumpkins.loc[new_pumpkins['Package'].str.contains('1 1/9'), 'Price'] = price/(1+1/9)
```

```python
new_pumpkins.loc[new_pumpkins['Package'].str.contains('1/2'), 'Price'] = price/(1/2)
```

```python
fig = plt.figure(figsize=(9,6))
ax = fig.gca()
new_pumpkins.boxplot('Price', by='Month', ax=ax)
plt.show()
```
