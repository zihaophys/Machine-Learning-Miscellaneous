```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.cluster import KMeans
```

```python
data, _ = make_classification(n_samples=1000, n_features=2, n_informative=2, n_redundant=0, n_clusters_per_class=1, random_state=4)
```

```python
data.shape
```

```python
fig = plt.figure(figsize=(9,6))
plt.xlim(-3, 2.5)
plt.ylim(-5,4.5)
plt.scatter(data.T[0], data.T[1])
plt.show()
```

```python
model = KMeans(n_clusters=2)
model.fit(data)
```

```python
result = model.predict(data)
```

```python
fig = plt.figure(figsize=(9,6))
data_0 = data[np.where(result==0)]
data_1 = data[np.where(result==1)]
plt.xlim(-3, 2.5)
plt.ylim(-5,4.5)
plt.scatter(data_0.T[0], data_0.T[1])
plt.scatter(data_1.T[0], data_1.T[1])
plt.show()
```

```python
from sklearn.cluster import DBSCAN
model = DBSCAN(eps=0.25, min_samples=10)
result = model.fit_predict(data)
fig = plt.figure(figsize=(9,6))
data_0 = data[np.where(result==0)]
data_1 = data[np.where(result==1)]
plt.xlim(-3, 2.5)
plt.ylim(-5,4.5)
for cluster in np.unique(result):
    cluster_data = data[np.where(result==cluster)]
    plt.scatter(cluster_data.T[0], cluster_data.T[1])
plt.show()
```

```python
np.unique(result)
```

```python
from sklearn.mixture import GaussianMixture
model = GaussianMixture(n_components=2)
result = model.fit_predict(data)
fig = plt.figure(figsize=(9,6))
data_0 = data[np.where(result==0)]
data_1 = data[np.where(result==1)]
plt.xlim(-3, 2.5)
plt.ylim(-5,4.5)
plt.scatter(data_0.T[0], data_0.T[1])
plt.scatter(data_1.T[0], data_1.T[1])
plt.show()
```

```python
from sklearn.cluster import Birch
model = Birch(threshold=0.03, n_clusters=2)
result = model.fit_predict(data)
fig = plt.figure(figsize=(9,6))
data_0 = data[np.where(result==0)]
data_1 = data[np.where(result==1)]
plt.xlim(-3, 2.5)
plt.ylim(-5,4.5)
plt.scatter(data_0.T[0], data_0.T[1])
plt.scatter(data_1.T[0], data_1.T[1])
plt.show()
```

```python
from sklearn.cluster import AffinityPropagation
model = AffinityPropagation(damping=0.95, random_state=0)
result = model.fit_predict(data)
fig = plt.figure(figsize=(9,6))
data_0 = data[np.where(result==0)]
data_1 = data[np.where(result==1)]
plt.xlim(-3, 2.5)
plt.ylim(-5,4.5)
for cluster in np.unique(result):
    cluster_data = data[np.where(result==cluster)]
    plt.scatter(cluster_data.T[0], cluster_data.T[1])
plt.show()
```
