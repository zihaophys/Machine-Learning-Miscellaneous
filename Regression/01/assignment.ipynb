## Linnerud dataset

```python
from sklearn import datasets
```

```python
X, y = datasets.load_linnerud(return_X_y=True, as_frame=True)
```

```python
X.describe()
```

```python
features = list(X.columns.values)
targets = list(y.columns.values)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
for col in features:
    fig, ax = plt.subplots(1, 2, figsize=(18, 6))
    data = X[col]
    ax[0].hist(data, bins=10)
    ax[0].axvline(data.mean(), linestyle='dashed', linewidth=2, color='red')
    ax[0].axvline(data.median(), linestyle='dashed', linewidth=2, color='green')
    ax[1].boxplot(data, vert=False)
    plt.suptitle(f'{col} distribution')
plt.show()
```

```python
for col in targets:
    fig, ax = plt.subplots(1, 2, figsize=(18, 6))
    data = y[col]
    ax[0].hist(data, bins=10)
    ax[0].axvline(data.mean(), linestyle='dashed', linewidth=2, color='red')
    ax[0].axvline(data.median(), linestyle='dashed', linewidth=2, color='green')
    ax[1].boxplot(data, vert=False)
    plt.suptitle(f'{col} distribution')
plt.show()
```

```python
for col in features:
    fig = plt.figure(figsize=(4,3))
    ax = fig.gca()
    plt.scatter(X[col], y['Waist'])
    ax.set_title(f'Waist vs {col}')
plt.show()
```

```python
X = X[y['Waist']< 40]
y = y[y['Waist']< 40]
```

```python
for col in features:
    fig = plt.figure(figsize=(4,3))
    ax = fig.gca()
    plt.scatter(X[col], y['Waist'])
    ax.set_title(f'Waist vs {col}')
plt.show()
```
